```python
data1 = data.dropna()      #since it does not have any null values data1 is equal to data
```

```python
data1.shape
```

```python
data1.describe()
```

```python
# Removes highly correlated columns
import numpy as np
threshold = 0.9

df_corr = data1.corr()
columns = np.full((df_corr.shape[0],), True, dtype=bool)
for i in range(df_corr.shape[0]):
    for j in range(i+1, df_corr.shape[0]):
        if df_corr.iloc[i,j] >= threshold:
            if columns[j]:
                columns[j] = False
selected_columns = data1.columns[columns]
selected_columns
data1 = data1[selected_columns]
data1.head()
```

```python
data1.shape
```

```python
X = data1.iloc[:, 1: ].values
y = data1.iloc[:,0]
```

```python
X
```

```python
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (60,60))
sns.heatmap(data1.corr(),cmap="YlGnBu", annot=True)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Logistic Regression as classifier

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
LRclassifier = LogisticRegression(max_iter = 300, random_state=0)
LRclassifier.fit(X_train, y_train)  
LRtrainpredictions = LRclassifier.predict(X_train)
LRtestpredictions = LRclassifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
LRtraincm = confusion_matrix(y_train, LRtrainpredictions)
LRtestcm = confusion_matrix(y_test, LRtestpredictions)
print('Logistic Regression Training Performance:\n ',LRtraincm)
print('Logistic Regression Testing Performance:\n', LRtestcm)
```

```python
# Confusion matrix based on training data
plot_confusion_matrix(LRclassifier, X_train, y_train)
plt.show()
```

```python
# Confusion matrix based on testing data
plot_confusion_matrix(LRclassifier, X_test, y_test)
plt.show()
```

```python
#Calculating Accuracy
LRtrainaccuracy = np.trace(LRtraincm)/np.sum(LRtraincm)
LRtestaccuracy = np.trace(LRtestcm)/np.sum(LRtestcm)
print('Logistic Regression Training Accuracy:', LRtrainaccuracy*100,'\b%\nLogistic Regression Testing Accuracy:',LRtestaccuracy*100,'\b%')
```

## LDA as classifier

```python
#LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
LDAclassifier = LDA()
LDAclassifier.fit(X_train, y_train)  
LDAtrainpredictions = LDAclassifier.predict(X_train)
LDAtestpredictions = LDAclassifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
LDAtraincm = confusion_matrix(y_train, LDAtrainpredictions)
LDAtestcm = confusion_matrix(y_test, LDAtestpredictions)
print('LDA Training Performance:\n ',LDAtraincm)
print('LDA Testing Performance:\n', LDAtestcm)
```

```python
# Confusion matrix based on training data
plot_confusion_matrix(LDAclassifier, X_train, y_train)
plt.show()
```

```python
# Confusion matrix based on testing data
plot_confusion_matrix(LDAclassifier, X_test, y_test)
plt.show()
```

```python
#Calculating Accuracy
LDAtrainaccuracy = np.trace(LDAtraincm)/np.sum(LDAtraincm)
LDAtestaccuracy = np.trace(LDAtestcm)/np.sum(LDAtestcm)
print('LDA Training Accuracy:', LDAtrainaccuracy*100,'\b%\nLDA Testing Accuracy:',LDAtestaccuracy*100,'\b%')
```

## QDA as classifier

```python
#QuadraticDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
QDAclassifier = QDA()
QDAclassifier.fit(X_train, y_train)  
QDAtrainpredictions = QDAclassifier.predict(X_train)
QDAtestpredictions = QDAclassifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
QDAtraincm = confusion_matrix(y_train, QDAtrainpredictions)
QDAtestcm = confusion_matrix(y_test, QDAtestpredictions)
print('QDA Training Performance:\n ',QDAtraincm)
print('QDA Testing Performance:\n', QDAtestcm)
```

```python
# Confusion matrix based on training data
plot_confusion_matrix(QDAclassifier, X_train, y_train)
plt.show()
```

```python
# Confusion matrix based on testing data
plot_confusion_matrix(QDAclassifier, X_test, y_test)
plt.show()
```

```python
#Calculating Accuracy
QDAtrainaccuracy = np.trace(QDAtraincm)/np.sum(QDAtraincm)
QDAtestaccuracy = np.trace(QDAtestcm)/np.sum(QDAtestcm)
print('QDA Training Accuracy:', QDAtrainaccuracy*100,'\b%\nQDA Testing Accuracy:',QDAtestaccuracy*100,'\b%')
```

## AUC and ROC Curve

```python
pred_prob1 = LRclassifier.predict_proba(X_test)
pred_prob2 = LDAclassifier.predict_proba(X_test)
pred_prob3 = QDAclassifier.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve
# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_test, pred_prob1[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_test, pred_prob2[:,1], pos_label=1)
fpr3, tpr3, thresh3 = roc_curve(y_test, pred_prob3[:,1], pos_label=1)
print(thresh1)
# roc curve for tpr = fpr
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score1 = roc_auc_score(y_test, pred_prob1[:,1])
auc_score2 = roc_auc_score(y_test, pred_prob2[:,1])
auc_score3 = roc_auc_score(y_test, pred_prob3[:,1])
print(auc_score1, auc_score2, auc_score3)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Logistic Regression')
plt.plot(fpr2, tpr2, linestyle='--',color='green', label='LDA')
plt.plot(fpr3, tpr3, linestyle='--',color='red', label='QDA')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.savefig('ROC',dpi=500)
plt.show()
```
